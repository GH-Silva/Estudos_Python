```python
import os
from glob import glob
import pyarrow.parquet as pq
import pyarrow as pa

# Caminho de origem dos arquivos .parquet
caminho_origem = r'C:\Users\GabrielHenriqueSilva\CAIXA Consórcio\Risco de Crédito e Antifraude - Documentos\BUSINESS INTELLIGENCE\INDICADORES\NOVA ESTRUTURA\INADIMPLÊNCIA\INADIMPLÊNCIA GERAL\Dados\Dados - Ramo de Atividade\PARQUET 1'

# Caminho de destino
caminho_destino = r'C:\Users\GabrielHenriqueSilva\CAIXA Consórcio\Risco de Crédito e Antifraude - Documentos\BUSINESS INTELLIGENCE\INDICADORES\NOVA ESTRUTURA\INADIMPLÊNCIA\INADIMPLÊNCIA GERAL\Dados\Dados - Ramo de Atividade\Principal\arquivo_unificado.parquet'

# Lista todos os arquivos .parquet na pasta de origem
arquivos = glob(os.path.join(caminho_origem, '*.parquet'))

# Inicializa variável para controle do schema
schema = None
writer = None

for idx, arquivo in enumerate(arquivos):
    print(f"Lendo e adicionando: {arquivo} ({idx + 1}/{len(arquivos)})")
    
    # Lê o arquivo como uma tabela Arrow
    tabela = pq.read_table(arquivo)
    
    # Inicializa o writer com o schema na primeira iteração
    if writer is None:
        schema = tabela.schema
        writer = pq.ParquetWriter(caminho_destino, schema)
    
    # Escreve a tabela no arquivo de destino
    writer.write_table(tabela)

# Fecha o writer para finalizar o arquivo
if writer:
    writer.close()

print(f"\n✅ Arquivo unificado criado com sucesso em:\n{caminho_destino}")
```
